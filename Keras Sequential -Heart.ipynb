```python
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
```

```python
from numpy import genfromtxt
my_data = genfromtxt('heart_failure_clinical_records_dataset.csv', delimiter=',')
```

```python
# split into input (X) and output (y) variables
X = my_data[:,0:12]
y = my_data[:,12]
```

```python
model = Sequential()
model.add(Dense(12, input_dim=12, activation='relu')) #input layer
model.add(Dense(24, activation='relu')) #hidden layer
model.add(Dense(1, activation='sigmoid')) #output layer
```

```python
# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=150, batch_size=20)
```

```python
_, accuracy = model.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))
```

```python
predictions = model.predict_classes(X)
```

```python
for i in range(10):
	print('%s => %d (expected %d)' % (X[i].tolist(), predictions[i], y[i]))
```
